```python
import matplotlib.pyplot as plt
```

```python
# Assume all pools have 100 tokens in each
def wombatExchangeRate(A, amount_swapped, tokens_in_pool):
    a = (1 + (A / ((tokens_in_pool - amount_swapped)/tokens_in_pool)**2))
    b = (1 + (A / ((tokens_in_pool + amount_swapped)/tokens_in_pool)**2))
    return b/ a


def uniswapExchangeRate(A, amount_swapped, tokens_in_pool):
    return (tokens_in_pool - amount_swapped) / (tokens_in_pool + amount_swapped)


```

```python
wombat = [wombatExchangeRate(.05, amount, 100) for amount in range(100)]
uniswap = [uniswapExchangeRate(.05, amount, 100) for amount in range(100)]

plt.plot([i for i in range(100)], wombat)
# plt.plot([i for i in range(100)], uniswap)
```
